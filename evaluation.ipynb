# Multimodal RAG Evaluation
This notebook measures retrieval performance (Hit Rate, MRR) for the system.

```python
import requests
import pandas as pd

API_URL = "http://localhost:8000"

# Define a small evaluation set: (query, expected_doc_id)
eval_set = [
    ("What is the revenue growth?", "report.pdf"),
    ("Show me the chart of units sold.", "report.pdf"),
]

results = []
for query, expected in eval_set:
    resp = requests.post(f"{API_URL}/query", json={"query": query, "n_results": 5})
    if resp.status_code == 200:
        sources = resp.json()["sources"]
        found_ids = [s["document_id"] for s in sources]
        hit = 1 if any(expected in fid for fid in found_ids) else 0
        results.append({"query": query, "hit": hit})

df = pd.DataFrame(results)
print(f"Hit Rate @ 5: {df['hit'].mean():.2f}")
```
